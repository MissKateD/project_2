 # Heart Attack Prediction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import pydotplus
from IPython.display import Image
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import tree
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
```

Read in Data

```python
heart_df = pd.read_csv("C:/Users/kd_84/Working/Project 2/project_2/data/heart.csv")
heart_df = heart_df.loc[:,['age', 'sex', 'cp', 'trestbps', 'chol', 'restecg', 'ca', 'thal', 'target']]
heart_df
```

Separate Features

```python
y = heart_df["target"]
X = heart_df.drop(columns="target")
```

Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=100)
```

Scale

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Learn

learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    model = GradientBoostingClassifier(
        n_estimators=100,
        learning_rate=learning_rate,
        max_features=2,
        max_depth=3,
        random_state=0)
    model.fit(X_train_scaled,y_train.ravel())
    print("Learning rate: ", learning_rate)

    print("Accuracy score (training): {0:.3f}".format(
        model.score(
            X_train_scaled,
            y_train.ravel())))
    print("Accuracy score (validation): {0:.3f}".format(
        model.score(
            X_test_scaled,
            y_test.ravel())))
    print()

Prune

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=100, max_depth=4)
model.fit(X_train_scaled, y_train)
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)
```

```python
print(accuracy_score(y_train, y_train_pred), round(accuracy_score(y_test, y_test_pred), 2))
```

```python
path = model.cost_complexity_pruning_path(X_train, y_train)
alphas = path['ccp_alphas']
alphas
```

```python
accuracy_train, accuracy_test = [], []
    
for i in alphas:
    model = DecisionTreeClassifier(ccp_alpha=i)
        
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
        
    accuracy_train.append(accuracy_score(y_train, y_train_pred))
    accuracy_test.append(accuracy_score(y_test, y_test_pred))
        
sns.set()
plt.figure(figsize = (14, 7))
sns.lineplot(y=accuracy_train, x=alphas, label="Train Accuracy")
sns.lineplot(y=accuracy_test, x=alphas, label="Test Accuracy")
plt.xticks(ticks=np.arange(0.00, 0.13, 0.01))
plt.show()                           
```

Peep the Tree

```python
model = DecisionTreeClassifier()
model = model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

acc_score = accuracy_score(y_test, predictions)
print(acc_score)
```

```python
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
import graphviz
dot_data = tree.export_graphviz(model, out_file=None, feature_names=X.columns, class_names=["0", "1"], filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)

Image(graph.create_png())
```
